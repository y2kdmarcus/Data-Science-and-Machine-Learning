Data Visualization

```python
#Line Chart
plt.plot(df['Month'], df['Sales'], marker = 'o', linestyle = '-', color='b', label = 'Data')
plt.title('Month vs Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

Why did you choose this graph?
- I chose this graph to visualize the trend of how the month affects the overall sales of the company.

What patterns or trends do you notice?
- A pattern seen here is a rising trend of sales with the highest being April to June.

What conclusions can be drawn from the data
- In a Philippine setting, a sales in these months could show how summer and back-to-school programs are in effect. Gadgets, Appliances, and Groceries could be bought around this time.

```python
# Bar Graph for Age vs Income
import io
df = pd.read_csv(io.BytesIO(uploaded['age_income_data.csv']))

plt.figure(figsize=(6, 4))
plt.bar(df['Age'], df['Income'], color='skyblue')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Why did you choose this graph?
- I chose bar graph to see the increase of salary as one ages.

What patterns or trends do you notice?
- The pattern being seen here is a direct trend, wherein as the age increase, salary also increases.

What conclusions can be drawn from the data
- Based from the data, we could reason as one ages with matched experience, their salary also increases.

```python
# Pie Chart for Revenue per Category
import io
df = pd.read_csv(io.BytesIO(uploaded['category_data.csv']))

plt.figure(figsize=(6, 4))
plt.pie(df['Revenue'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Revenue per Category')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

Why did you choose this graph?
- I chose pie graph to see how the total revenue of the company is being divided per category.

What patterns or trends do you notice?
- We can see that electronics is the highest revenue for the company followed by the clothing

What conclusions can be drawn from the data
- We can conclude that consumers buy electronics such as gadgets and appliance in the store, compared to other categories.

```python
# Pie Chart for Customer Satisfaction
import io

df = pd.read_csv(io.BytesIO(uploaded['ratings_data.csv']))

plt.figure(figsize=(6, 4))
plt.pie(df['Count'], labels=df['Satisfaction_Level'], autopct='%1.1f%%', startangle=90)
plt.title('Customer Satisfaction')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

Why did you choose this graph?
- I chose pie graph to check how many customers are satisfied with the service the company offers.

What patterns or trends do you notice?
- We can see that most consumers are fine or neutral with the service they are providing.

What conclusions can be drawn from the data
- The service and overall satisfaction of the consumers are neutral. Certain imporvements needs to be made in order to decrease the dissatisfaction rate, increase satisfaction, and transfer the neutral to satisfaction.
