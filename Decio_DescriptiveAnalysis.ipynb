```python
import pandas as pd

from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['Decio_Dataset.csv']))
print(f"DataFrame shape before removing rows: {df.shape}")

# Keep rows from index 0 up to (but not including) index 46
df = df.iloc[:45]

print(f"DataFrame shape after removing rows: {df.shape}")
df
```

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Measures of Central Tendency
print("Mean\n", df.mean(numeric_only=True))
print("\nMedian\n", df.median(numeric_only=True))
print("\nMode\n", df.mode().iloc[0])
print("\nStandard Deviation:\n", df.std(numeric_only=True))
print("\nVariance:\n", df.var(numeric_only=True))
```

The mean resutlts have shown that the whole world averages 129.09 litres of beer, which is considered average. While spirit drinks like gin or vodka are being consumed on average by 91.522. Also, wine servings are also considered low on average compared to largest wine consumers like Australia. Lastly, the total litres on average is 5.506 which is on the lower bracket especially when compared to several countries in Europe. The median are considered on the lower end as same as the mean. This could mean that there are many outliers in the higher mark. The mode shows that it is below average in all categories.

```python
# Measures of Variability
print("\nRange")
for col in ['beer_servings','spirit_servings','wine_servings', 'total_litres_of_pure_alcohol']:
 print(col, df[col].max() - df[col].min())

# IQR
q1 = df.quantile(0.25, numeric_only=True)
q3 = df.quantile(0.75, numeric_only=True)
print("\nIQR:\n", q3 - q1)

# Skewness & Kurtosis
print("\nSkewness:\n", df.skew(numeric_only=True))
print("\nKurtosis:\n", df.kurtosis(numeric_only=True))

# Tables and Graphs
df[['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']].hist(bins=20)
plt.suptitle("Score Distributions")
plt.show()

# Boxplot for Total Liters of Alcohol
plt.figure() # Create a new figure for the boxplot
# Drop NaN values from the column before plotting to ensure clean data
data_for_boxplot = df['total_litres_of_pure_alcohol'].dropna()

if not data_for_boxplot.empty and pd.api.types.is_numeric_dtype(data_for_boxplot):
    plt.boxplot(data_for_boxplot)
    plt.title("Total Liters of Alcohol Boxplot")
    plt.ylabel("Litres of Pure Alcohol")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```
